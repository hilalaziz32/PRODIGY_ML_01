```python
from re import L
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df=pd.read_csv('train.csv')

X=df[['BedroomAbvGr','FullBath','GrLivArea']]
Y=df['SalePrice']

X=X.fillna(X.median())

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(scaled_X, Y, test_size=0.33, random_state=42)

model=LinearRegression()
model.fit(X_train,Y_train)

Y_pred=model.predict(X_test)
print(f'Mean Squared Error: {mean_squared_error(Y_test, Y_pred)}')
print(f'R² Score: {r2_score(Y_test, Y_pred)}')

cv_scores = cross_val_score(model, scaled_X,Y, cv=5)
print(f'Cross-Validation Scores: {cv_scores}')
print(f'Average CV Score: {np.mean(cv_scores)}')


df=pd.read_csv('test.csv')
X_test=df[['BedroomAbvGr','FullBath','GrLivArea']]
X_test=X_test.fillna(X_test.median())
scaled_X_test = scaler.transform(X_test)

price_pred=model.predict(scaled_X_test)
prices = pd.DataFrame({
    'Id': df['Id'],
    'SalePrice': price_pred
})

prices.to_csv('submission.csv', index=False)
print(prices.head().to_string(index=False))
```
